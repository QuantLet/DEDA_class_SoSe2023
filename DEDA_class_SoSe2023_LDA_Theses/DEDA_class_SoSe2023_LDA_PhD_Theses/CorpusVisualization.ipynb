```python
import matplotlib.pyplot as pPlot
from wordcloud import WordCloud, STOPWORDS
import numpy as np
from PIL import Image


from sklearn.feature_extraction.text import TfidfVectorizer
import umap
import umap.plot
```

```python
pdf_out_folder_path = './full_text'
text_path = f'{pdf_out_folder_path}/cleaned/cleaned_text/all.txt'
```

## Creating the word cloud

```python
dataset = open(text_path, "r").read()

def create_word_cloud(string):
   cloud = WordCloud(background_color = "white", max_words = 200, stopwords = set(STOPWORDS))
   cloud.generate(string)
   cloud.to_file("wordCloud.png")

create_word_cloud(dataset)
```

## Creating the UMAP

```python
dataset = open(text_path, "r").read()

# Load the text data
corpus = dataset.split(" ")

tfidf = TfidfVectorizer()
docs = tfidf.fit_transform(corpus)
```

```python
mapper = umap.UMAP().fit(docs)
umap.plot.points(mapper)
```
