```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def make_plot(path, validation_set, x_axis_column, y_axis_column):
    df = pd.read_csv(path)

    df_75 = df[df['Validation_Set'] == validation_set]
    df_75 = df_75.groupby(x_axis_column)\
      .apply(lambda x: x.loc[x[y_axis_column]==x[y_axis_column].max()])

    plt.plot(df_75[x_axis_column], df_75[y_axis_column])
    plt.title(f'{y_axis_column} Score')
    plt.xlabel(x_axis_column)
    plt.ylabel(y_axis_column)

    plt.savefig(f"./results/{y_axis_column}_scores_corpus.png", dpi = 300, transparent = True)
    plt.show()
    plt.close()
```

```python
path = './results/lda_tuning_results_tfidf.csv'
```

```python
df = pd.read_csv(path)
df = df[df['Validation_Set'] == "100% Corpus"]
print(df.head())
max_id = df['Coherence'].idxmax()
print(df.loc[max_id])
```

<h1>75% Corpus</h1>

```python
make_plot(path, "75% Corpus", "Topics", "Coherence")
```

```python
make_plot(path, "75% Corpus", "Alpha", "Coherence")
```

```python
make_plot(path, "75% Corpus", "Topics", "Perplexity")
```
